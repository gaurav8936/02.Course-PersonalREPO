<p>Note that we can change file &amp; folder permissions: each has an owner, can be part of a group, and can accessed by "the world" (or "other").</p>
<p>An <i style="color:red">owner</i> is the person who creates the file and has control over the <b>read</b>, <b>write</b>, and <b>execute</b> permissions on that file/directory. <p>
    <p>Usually a folder/file may be shared, say by everyone in the sales department, and so form a <i style="color:red">group</i>.  The owner can extend to others in the group the same permissions.
        </p>
<p>Finally, we may want to share a file with the <i style="color:red">world</i> (or other), as we do with web pages.</p>
<p>Typically, our files by default may have permissions of <b>0644</b> for the owner to read/write, and others can only read the file. [Or <b>0755</b> (for webpages).]<br />If the file is a script or program, we may want to run (execute) the file - but not let others (like hackers) do so.  Let&rsquo;s look at the file permissions and how we can issue commands to the operating system in both the terminal window and Jupyter Notebook.</p>

<hr />
<b>Communicating with the OS</b><br />To communicate between the OS and Notebook, use !.  A similar command is % (or %%), called a "magic function."  To invoke the Unix terminal use the magic function <span style="font-family:Courier">%%bash</span>. For Windows, you could try <span style="font-family:Courier">%%cmd</span>.

```python
%%bash
head welcome.txt
date
```

```python
%%bash
clear  # clear the screen

history # show the list of previous commands (using the up/down arrow - very helpful)

date # shows current date and time
uptime # how long the system has been on...

whoami # identity crisis averted - username
id 
who 
```

```python
%%bash
ls # "list show" - the files and directories in the cwd (current working directory)
pwd  # "print" the working directory
```

<hr />
<p>These commands affect files - they&rsquo;re important to know.  Remember that the space is a significant code - it separates the command from inputs (parameters and switches) to that command.
<ul><li><code>ls</code> - list show - all files &amp; directories</li>
    <li><code>cp</code> - copy files (from source to destination, e.g., <code>cp mya.txt myb.txt</code></li>
    <li><code>rm</code> - remove files and directories, e.g., rm myb.txt</li>
    <li><code>rm</code> - rename or move (from ... to ...)</li>
    <li><code>chmod</code> - change mode - changes file/directory access permissions</li>
    <li><code>chown</code> - change the ownership of a file</li>
    <li><code>cat</code> - concatenate a couple of files and shows in the stdout</li>
    <li><code>head</code> - show the first lines of a test file, default is 10</li>
    <li><code>tail</code> - show the bottom lines, 10 default</li>
    <li><code>grep</code> - search for patterns in text files.</li>
    <li><code>cd</code> - change directory; e.g., <code>cd ..</code> means go up to the parent directory</li>
    <li><code>ln</code> - make links and symlinks to files/directories</li>
    <li><code>mkdir</code> - make a new directory, e.g., mkdir w200homeworks</li>
    <li><code>rmdir</code> - remove a directory, e.g., rmdir myfolder</li>
    <li><code>sudo</code> - change temporarily to superuser (use with great care!); some shells use su instead of sudo</li>
    <li><code>ps</code> - show the running processes (tasks, jobs) - the PID (process ID), input source (tty), time running, and the command used to start the process</li>
    <li><code>top</code> - show tasks and system status</li>
    <li><code>kill</code> - kill a process - but you need to know the process ID</li>
    <li><code>&gt;</code> - a "redirector" - to redirect the output from the stdout to something else, like a file.  E.g., <code>ls &gt; listOfMyFiles.txt</code></li>
    <li><code>|</code> - a "pipe" - a Unix communication channel - used often to take the results of one command and send them (pipe them) to another command</li>
</ul>
</p>  

<hr /><h2>Some network commands</h2>
<p>There is always only one active <b>stdin</b> (standard input, default is keyboard]), <b>stdout</b> (standard output, default is the monitor), and <b>stderr</b> (standard error, an error log file).  We redirect these when reading/writing to files and when communicating over the net.  Contemporary programming languages allow switching easy between local and network systems, by identifying the <b>protocol</b> or via some bridge.  Compare accessing files using different protocols <code>ftp:</code> (file transfer protocol, very insecure), <code>sftp</code> (encrypted ftp), <code>wget</code> (download files from remote uservers using http/https or ftp, <code>ssh</code> (secure socket shell, enrypted remote access client), <code>http://</code>, <code>https://</code>, <code>file://</code> and so on.  We can't review file transfers here but let's look at some network-related commands.</p>
<p>Below are some network commands used to reveal info about the Unix system: notice hostname will likely include "local" in the name (as opposed to an IP address that you'll see in the ip and netstat commands).  <code>uname</code> prints the Unix system info - on a Mac you'll see "Darwin".  When distributing code it's useful to determine the users' OS and install any dependencies (libraries) for your code.</p>

```python
%%bash
hostname 
w
uname
```

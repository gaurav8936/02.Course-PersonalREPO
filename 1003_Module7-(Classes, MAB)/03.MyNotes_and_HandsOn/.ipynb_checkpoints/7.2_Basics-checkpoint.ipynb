### Data Attributes

```python
# defining attributes from the outside (this is not recommended)
```

```python
Drone.power_system = 'Battery'

# this just adds a new attribute with a constant value of 'Bettery'
# this is added to all objects form the class
```

```python
# this is a class attribute
```

```python
# this now applies to all drones
```

```python
# for the instance

print(d1.power_system)
```

```python
# for the instance

d2.power_system
```

```python
# for the class

Drone.power_system
```

# Difference between REPR and STR

```python
class Animal():
    
    def __init__(self, name):
        self.name = name
        
    def __str__(self):
        return f"I am a {self.name}."
```

```python
b1 = Animal('Deer')
```

```python
b1
```

```python
# you dont see the description as with repr
```

```python
print(b1)
```

### REPR does both things (just a2 and also print(a2)) -- whereas STR just does one (print(a2))

---

```python
class Animal():
    
    def __init__(self, name):
        self.name = name
        
    def __repr__(self):
        return f"I am a {self.name}."

    def __str__(self):
        print('Calling repr')
        return self.__repr__()
    
    # this is just a case where etr is calling repr
```

```python
c1 = Animal('Heer')
```

```python
c1
```

```python
print(c1)
```

```python
class Animal():
    
    def __init__(self, name):
        self.name = name
        
    def __repr__(self):
        return f"I am a {self.name}."

    def __str__(self):
        print('Calling repr')
        return self.__repr__()

class Zoo():
    
    def __init__(self):
        self.animals = []
    
    def add_animal(self, animal):
        self.animals.append(animal)
```

```python
d1 = Animal('d1')
d2 = Animal('d2')
d3 = Animal('d3')
d4 = Animal('d4')
```

```python
zoo.add_animal(d1)
zoo.add_animal(d2)
zoo.add_animal(d3)
zoo.add_animal(d4)
```

```python
zoo.animals
```

```python
print(zoo.animals)
```

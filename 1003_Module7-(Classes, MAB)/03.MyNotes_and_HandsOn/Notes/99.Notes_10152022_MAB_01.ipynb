```python
import numpy as np

class Bandit():
    
    def __init__(self):
        
        self.__probability = np.random.uniform()
    
    def __repr__(self):
        
        return 'this is a bandit with a secret probability of ' + str(self.__probability)
        # REPR can contain only strings (no integers or floats) -- has to be descriptive 
    
    def pull(self):
        
        # this should return a true or false depending on whether a pull was successful or not
        # true - success
        # false - failure
        
#        return 'this returns a success as true or false failure', np.random.choice((True, False, [self.__probability, 1-self.__probability]))
#        return np.random.choice((True, False), [self.__probability, 1-self.__probability])
        return np.random.choice((True, False), p=[self.__probability, 1-self.__probability])
```

```python
b1 = Bandit()
b2 = Bandit()
```

```python
print(b1)
print(b2)
```

```python
b1.pull()
```

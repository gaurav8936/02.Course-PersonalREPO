```python
import numpy as np
import matplotlib.pyplot as plt


class Bandit():
    
    def __init__(self):
        
        self.__probability = np.random.uniform()
    
    
    def __repr__(self):
        
        return "This is a bandit with a secret probability of " + str(self.__probability)
    
    def pull(self):
        
        a = (True, False)
        return np.random.choice(a, p = [self.__probability, 1-self.__probability])
```

```python
import matplotlib.pyplot as plt

class Game():
    
    def __init__(self, num_moves):
        self.num_moves = num_moves
        self.score = 0
        self.history = []
        self.best_probability = 0
    
    def moves_remaining(self):
        # yes or no
        return self.num_moves > len(self.history)
    
    def plot_history(self):
        plt.plot(np.cumsum(self.history))
        plt.title('Reward History')
        plt.xlabel('Move Number')
        plt.ylabel('Total Reward')
        plt.plot([0, len(self.history)], [0, len(self.history) * self.best_probability], color = 'r', linestyle = '--', linewidth=1) # added new

        
```

```python
class Bandit:
    
    def __init__(self, game):
        self.__probability = np.random.uniform()

        self.game = game # initialize the game
        if self.__probability > self.game.best_probability:
            self.game.best_probability = self.__probability
        
    def __repr__(self):
        return "this is a bandit object with secret probability "  + str(self.__probability)
    
    def pull(self):
        
        if not self.game.moves_remaining():
            raise Exception("Bandit pull called with no moves remaining")
            
        result = np.random.choice((True, False), p=[self.__probability, 1-self.__probability])

        self.game.score += result
        self.game.history.append(result)
```

```python
class RandomSolver():
    
    """
    Represents the player playing the game
    """
    
    def __init__(self, bandits):
        
        self.bandits = bandits
        
        # we initialize the player along with the bandit (else, there is no reason to exist)
    
    def move(self):
        
        # select bandit at random
        # pull method
        
        chosen = np.random.randint(0, len(bandits))
        print('choosing the bandit number', chosen)
        
        # pull and get result
        
        return bandits[chosen].pull()
```

# Revising for best probability

# ie if the player had always made the most logical decision?


```python
# Epsisolon Greedy Algorithm

#epsiloon is a small probability (say, 1% or .1%)
# with probability epsilon we do exactly as we did before
# choose a bandit at random
# randomly break ties
# otherwise, choose the one that has the best result thus far
```

```python
class EpsilonGreedySolver():
    
    """
    Represents the player playing the game
    """
    
    def __init__(self, bandits, epsilon = 0.01):
        
        self.bandits = bandits
        self.epsilon = epsilon
        self.trials_per_bandit = [0 for i in range(len(bandits))]
        self.rewards_per_bandit = [0 for i in range(len(bandits))]
        self.score_per_bandit = [0 for i in range(len(bandits))]
                                  
        
        # we initialize the player along with the bandit (else, there is no reason to exist)
    
    def move(self):
        
        # select bandit at random
        # pull method
        
        # this is the exploration idea
        if np.random.uniform() < self.epsilon: # this will be true exactly of probability epsilon (property of uniform
                                            # distributions)
        
            chosen = np.random.randint(0, len(bandits))
        
       
        # then exploitation
        
        else:
            # we need to add logic to find out the best bandit so far
            # how many time we tried -- how many rewards
            # keep track of win ration
            
            max_score = 0
            best = [] # could be more than one
            for i in range(len(self.bandits)):
                if self.score_per_bandit[i] == max_score:
                    best.append(i)
                    
                if self.score_per_bandit[i] > max_score:
                    best = [i]
                    max_score = self.score_per_bandit[i]
                
                chosen = np.random.choice(best)
            
        result = bandits[chosen].pull()
        self.trials_per_bandit[chosen] += 1
        self.rewards_per_bandit[chosen] += result # if reward is true
        self.score_per_bandit[chosen] = self.rewards_per_bandit[chosen] / self.trials_per_bandit[chosen]
        
        #print('choosing the bandit number', chosen)
        
        # pull and get result
        
        return result
```

```python
game = Game(300)
bandits = [Bandit(game) for i in range(10)]
solver = EpsilonGreedySolver(bandits, 0.01)
```

```python
for i in range(game.num_moves):
    solver.move()
game.plot_history()
```

```python
import matplotlib.pyplot as plt

class Game:
    
    """
    This is a class to represent the overall rules of the game
    """
    
    def __init__(self, num_moves):
        
        self.num_moves = num_moves # store as an instance attr
        
        self.score = 0
        
        self.history = []
        
        self.best_probability = 0 # added new
        
    def moves_remaining(self):
        
        return self.num_moves > len(self.history)
    
    def plot_history(self):
        
        plt.plot(np.cumsum(self.history))
        plt.title('Reward History')
        plt.xlabel('move number')
        plt.ylabel('total reward')
        plt.plot([0, len(self.history)], [0, len(self.history) * self.best_probability], color = 'r', linestyle = '--', linewidth=1) # added new
```

```python
import numpy as np

class Bandit:
    
    """
    
    Class to represent a slot machine with a secret probability of reward
    """
    
    def __init__(self, game):
        self.__probability = np.random.uniform()
        
        self.game = game

        if self.__probability > self.game.best_probability: # added
            self.game.best_probability = self.__probability # added
        
    def __repr__(self):
        return "this is a bandit object with secret probability "  + str(self.__probability)
    
    def pull(self):
        
        """
        
        returns true representing a reward or false representing no reward
        """
        
        if not self.game.moves_remaining():
            raise Exception ("Bandit pull called with no moves remaining")
        
        result = np.random.choice((True, False), p=[self.__probability, 1-self.__probability])
        
        self.game.history.append(result)
        
        self.game.score += result
        
        return result # not to break previous functinality
```

```python
game = Game(300)

bandits = [Bandit(game) for i in range(10)]

s = EpsilonGreedySolver(bandits, .01)

for i in range(game.num_moves):
    s.move()

game.plot_history()
```

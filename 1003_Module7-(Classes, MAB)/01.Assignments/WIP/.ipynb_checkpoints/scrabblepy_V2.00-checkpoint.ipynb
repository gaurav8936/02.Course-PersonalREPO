```python
def convert_str_to_list(user_input):
    input_list = []
    for letter in user_input.lower():
        input_list.append(letter)
        input_list.sort() # not sure this is needed
    return (input_list)
    print(input_list)
```

```python
input_list = convert_str_to_list(user_input)
print(input_list)
```

```python
# truncate the master dictionary to words less than or equal to the input string

def trunc_masterdict_to_input(masterdict, input_list):
    
    masterdict_trunc = []
#    print(masterdict_matched)
    
    for item in masterdict:
        if len(item) <= len(input_list):
#            print(len(item))
#            print(len(input_list))
            masterdict_trunc.append(item)
     
    return masterdict_trunc
```

```python
trunc_masterdict_to_input(masterdict, input_list)
```

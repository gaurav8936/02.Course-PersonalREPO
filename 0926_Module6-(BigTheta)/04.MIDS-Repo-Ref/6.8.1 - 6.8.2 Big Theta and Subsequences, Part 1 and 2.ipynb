## Big-Î˜ and Subsequences

```python
import numpy as np
import matplotlib.pyplot as plt

price_changes = np.random.normal(0,1,size=50)
plt.plot(price_changes)
plt.title("Daily Price Change")
```

#### Problem: find the subarray with the highest total price change

```python
best = price_changes[0]
for start in range(len(price_changes)):
    for end in range(start+1, len(price_changes)+1):
        if sum(price_changes[start:end]) > best:
            best = sum(price_changes[start:end])

print(best)
```

$\Theta(n^2)$

```python
cumulative_price_change = [0]
for change in price_changes:
    cumulative_price_change.append(cumulative_price_change[-1] + change)
                                   
plt.plot(cumulative_price_change)
```

```python
max(cumulative_price_change) - min(cumulative_price_change)
```

$\Theta(n)$

### Refinement: Remember min and max along the way

```python
cumulative_price_change = [0]
min_price = 0
max_price = 0
for change in price_changes:
    cumulative_price_change.append(cumulative_price_change[-1] + change)
    min_price = min(min_price, cumulative_price_change[-1])
    max_price = max(max_price, cumulative_price_change[-1])
                                   
max_price - min_price
```

$\Theta(n)$

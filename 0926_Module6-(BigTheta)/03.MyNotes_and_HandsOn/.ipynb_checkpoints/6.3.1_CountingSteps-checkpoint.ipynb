```python
n = 5

for i in range(n):
    print('this is: ', i)
    for j in range(i):
        print(j)
```

```python
# first time through i = 0
# print does not execute
# we are going through the outer loop once
```

```python
# ssecond time i = 1
# 3 in total (1 for outer, 1 for inner, 1 for print)
```

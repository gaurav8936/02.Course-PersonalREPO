```python
for i in range(5):
    print(i)
```

```python
# looks like BigO (n)

# this will take the same amount of time regardless of what we enter

# this is BigO (1) and not BigO (n) as thought earlier

# as you can see, there is no relationship between n and range(5)

##### this is why this is BigO (1)
```

---

```python
for i in range(n):
    print(i)
```

```python
# here the range is n
# so, there will n different print stateemnts
```

```python
# this is bigO (n)
# teh time increases along with n
```

---

```python
for i in range(n):
    for j in range(n):
        print("*", end="")
    print()
```

### add all the stars -- and you get n^2

```python
# the only thing that really matters is the innermost print

# this has to execute more than any other statement (eveyr time we pass through for j)
```

```python
# the outer print will execute n times

# the inner print loop will execute n**2 times
```

```python
# thetaO (n**2)
```

---

### the j loop now only goes up to i (not to n)

```python
for i in range(n):
    for j in range(i):
        print("*", end="")
    print()
```

```python
# could this be lesser than O(n**2)
```

```python
# answer is NO
```

```python
# we are printing half of the square
```

```python
# but going back to the growth formular

# we can take out constant factors (like 1/2 in this case)
```

```python
# hence, this is bigO(n**2)
```

### the bottomline is that we aget n^2/4 and we just ignote the constant (1/4) -- which makes is n^2

```python
# the smaller square has width of 1/2 of n and height of 1/2 n

# again this is a constant which can be ignore
```

```python
import time
```

```python
times = []
```

```python
sizes = list(range(1,1000))

for n in sizes:
    price_changes = np.random.uniform(0,100, size = n)
    # this is the array of random numbers 
    
    start_time = time.time()
    
    max(price_changes)
    
    end_time = time.time()
    
    times.append(end_time - start_time)
```

```python
times
```

```python
import matplotlib.pyplot as plt

```

```python
plt.scatter(x = sizes, y = times)
```

```python
plt.scatter(x = sizes, y = times)

plt.title("Execution times for max experiment")
```

```python
# under the hood, the way python implements the max function is linear time

# you can see the line

# this is an example of bigO(n)
```

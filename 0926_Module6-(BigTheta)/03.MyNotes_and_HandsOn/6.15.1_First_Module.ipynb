# Importing an external Module

## import a specific function directly from the external module

```python
from weatherman import get_report

# we imported just the specific function

print(get_report())

# we can ust write get_report 
```

---

## Import Module Name to avoid naming conflicts

```python
import weatherman

# imports the full module

print(weatherman.get_report())
```

---

## Import only what you want from the module by renamng the function

```python
from weatherman import get_report as do_it

# we import only what we need from the weatherman module (ie the get_report) and rename the function

print(do_it())
```

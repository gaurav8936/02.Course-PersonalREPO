# Namespaces

```python
# we tend to use the same variable names in different functions
```

```python
def add_one(x):
    x += 1
    print('in the function, x = ',x)
    return x

x = 3
result = add_one(x)
print('the function returned ', result)
print('Outside the function, x =', x)
```

```python
# outside the function, the value of x does not change
```

### there isn't just one x variable

### difference between global vs local variables

##### these are like two totally different variables

## Defining a local variable inside a function

```python
def add_one(x):
    y = x + 1
    print('in the function, y = ',y)
    return x

x = 3
y = 3
result = add_one(x)
print('the function returned ', result)
print('Outside the function, y =', y)
```

##### we have a local Y and a global Y

# Accessing Global Variables

```python
def add_y(x):
    ans = x + y
    # notice there is not local y
    # x is a local variable
    print("In the function, y", y)
    return ans
    

y = 5
result = add_y(3)
print("the function returned ", result)
print("Outside the function, y ", y)
```

```python
# the local function inherited the global variable y

# first checks in the local namsepace

# since there is no y, it goes up the STACK and finds the global namespace
```

### changing values of global variables

```python
def change_y():
    y = y + 1
    print("Inside the function, y", y)

y = 5
change_y()
print("Outside the function, y", y)
```

#### unbound local error

```python
# STEPS IN CALLING A FUNCTION

# 1. Create a new local namespace

# 2. For each parameter, create a new local variable and bind to argument being passed

# 3. For each assignment statement, if the name being assigned to is 
#    not in the local namespacek create a new local variable

# 4. Execute the statements
```

```python
# in the above example, y has no value BEFORE it is being assigned (ie y = y+1)
```

<font color = red>This also prevents the accidental change of global variable values from inside a local variable

## to change the global variable y

```python
def change_y():
    global y
    y = y + 1
    print("Inside the function, y", y)

y = 5
change_y()
print("Outside the function, y", y)
```

<font size = 4, color = red>Adding the 'global y' allows us to change the global value as well

### Treat this as an exception -- use GLOBAL VARIABLES SPARINGLY

#### as this goes against the grain of python & makes the code fragile

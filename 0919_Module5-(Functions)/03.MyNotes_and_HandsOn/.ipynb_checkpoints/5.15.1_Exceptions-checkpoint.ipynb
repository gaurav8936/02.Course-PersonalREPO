# TRY

```python

try:
    x = float(input("enter a number "))
    print("the reciprocal of the number is ", 1/x)
except:
    print('you did not enter a valid number!')
```

```python
try:

    x = float(input("enter a number "))
    print("the reciprocal of the number is ", 1/x)

except ValueError:
    print("You did not enter a number")

except ZeroDivisionError:
    print("Zero does not have a reciprocal")

except:
    print("Something else went wrong")
```

# Creating an Exception

```python
def sell(item, quantity, inventory):
    if item not in inventory:
        raise Exception(str(item) + " does not appear in inventory")
    
    q = inventory(item)
    
    if q < quantity:
        raise Exception("Invenotry does not contain enough items to sell")
    
    inventory[item] = q - quantity
    
inventory = {"oranges" : 10, "apple": 5, "plums": 3}
```

```python
sell("oranges", 15, inventory)
```

---

# Print the Error Message (rather than in stack trace)

```python
try:
    sell("oranges", 15, inventory)

except Exception as e:
    
    print("Could not complete sales: ", str(e))
```

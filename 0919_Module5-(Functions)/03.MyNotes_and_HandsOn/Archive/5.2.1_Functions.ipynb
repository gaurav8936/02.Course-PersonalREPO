### Calculating SQUARE ROOT

```python
# Newton's method for finding square roots

"""
writing a function to calculate teh square root based
on Newtons' method
"""

def sqrt(x, epsilon):
    ans = 1
    num_guesses = 0

    while abs(x/ans - ans) > epsilon:
        ans = (x/ans + ans) / 2
        num_guesses += 1
    
    return ans
```

```python
print(sqrt(2, .00001))
```

```python
# calculating the fourth root

square_one = (sqrt(4, .00001))
print(square_one)

square_two = (sqrt(square_one, .00001))
print(square_two)
```

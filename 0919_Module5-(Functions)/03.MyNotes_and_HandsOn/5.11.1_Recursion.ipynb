# Recursion (function calling itself)

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

# fuction calling itself
    
```

```python
factorial(1)
```

```python
factorial(4)
```

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

# fuction calling itself

print(factorial(3))
```

### multiple copies of factorial (function) -- as each function has its own namespace

#### for a recursive algorithm we need 2 things

##### 1. BASECASE (simply case - say for n = 1)

##### 2. RECURSIVE RULE (A mathematical way to write the rule)

factorial(n) = n * (n-1) * (n-2) * ... 

= n * factorial(n-1) --- which is a mathematical fact

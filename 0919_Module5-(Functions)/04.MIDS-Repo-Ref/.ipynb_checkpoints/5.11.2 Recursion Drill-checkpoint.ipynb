If num is prime, return "Prime!"  If num is not prime, return "Not Prime!".

```python
# Your code here

# 5
# 5 / 2
# 5 / 3
# 5 / 4

x = 2

def checkPrime(y):
    x = 2
    while y > 2:
        
        if y % x == 0:
            #print('number is not a prime', y)
            return(y, 'is not a prime')
            break
        else:
            # this assumes y is not divisible by x
            # inrement x by 1
            x = x + 1
            
            # check if x == y 
            # if it gets to this point, then it is a prime
            if x == y:
                #print(y, 'is a prime')
                return(y, 'a prime')
                break
#    print(y)
    
# print(checkPrime(9))

# print(checkPrime(5))
```

#### solution (course)

```python
def checkPrime(num):
    
    x = 2
    while x < num and num % x != 0:
        x += 1
    
    if x == num:
        return "Prime!"
    else:
        return "Not Prime!"
```

```python
print(checkPrime(4))
```

```python
print(checkPrime(7))
print(checkPrime(10))
```

```python
def checkPrime(num):
    
    x = 2
    while x < num and num % x != 0:
        x += 1
    
    if x == num:
        return [num]
    else: 
        return "Not Prime!"
```

```python
def checkPrime(num):
    
    x = 2
    while x < num and num % x != 0:
        x += 1
    
    if x == num:
        return [num]
    else: 
        return checkPrime(x) + checkPrime(int(num / x))
```

```python
print(checkPrime(7))
print(checkPrime(10))
print(checkPrime(25))
print(checkPrime(98))
```

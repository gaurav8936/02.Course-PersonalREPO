# Magic Methods

```python
class Card:
    
    def __init__(self, value, suit):
        self.value = value
        self.suit = suit
    
    #equality string
    def __eq__(self, other):
        if self.value == other.value:
            return True
        else:
            return False
    
    # less than    
    def __lt__(self, other):
        
        if self.value < other.value:
            return True
        else:
            return False
    
    # greater than    
    def __gt__(self, other):
        
        if self.value > other.value:
            return True
        else:
            return False
            
```

```python
class Card:
    
    def __init__(self, value, suit):
        self.value = value
        self.suit = suit
    
    #equality string
    def __eq__(self, other):
        if self.value == other.value:
            return True
        else:
            return False
    
    # less than    
    def __lt__(self, other):
        
        if self.value < other.value:
            return True
        else:
            return False
    
    # greater than    
    def __gt__(self, other):
        
        if self.value > other.value:
            return True
        else:
            return False
    
    # repr
    
    def __repr__(self):
        return "%i of %s " % (self.value, self.suit)
```

```python
print(Card(10, 'hearts'))
```

```python
cards = []

for suit in ('hearts', 'spades', 'diamonds', 'clubs'):
    for value in range(0,13):
        cards.append(Card(value, suit))
```

# Creating a new VECTOR Class

```python
# immutable lists of numbers with certain properties
```

<font size = 5, color = red> Vector : They're simply immutable lists of numbers with certain mathematical properties. So one of these properties is that you can take a dot product or inner product of two vectors. This product is a summation of all the values in the vector

```python
class Vector:
    
    def __init__(self, numbers):
        self.__numbers = numbers
    
    @property
    def numbers(self):
        return self.__numbers
    
    def dot(self, other):
        values = []
        
        if len(self.__numbers) != len(other.numbers):
            return "Cannot multiple different size vectors"
        
        for x in range(len(self.__numbers)):
            this_num = self.__numbers[x]
            other_num = other.numbers[x]
            values.append(this_num * other_num)
        return sum(values)
    
    def __mul__(self, other):
        return self.dot(other)
```

```python
a = Vector([1,10,4])
```

```python
b = Vector([2, -1, 5])
```

```python
print(a*b)
print(a.dot(b))
```

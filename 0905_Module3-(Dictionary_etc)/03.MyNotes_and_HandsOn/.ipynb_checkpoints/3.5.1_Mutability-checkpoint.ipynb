# Deep Copy

```python
from copy import deepcopy
```

```python
a11 = [1,2,[3,4]]
```

```python
a11
```

```python
b11 = a11.copy()
```

```python
b11
```

```python
c11 = deepcopy(a11)
```

```python
c11
```

```python
a11[2].append("Paul")
```

```python
a11
```

```python
b11
```

```python
c11
```

## Note that c11 retains the original a11 and is not mutated (after a11 was changed)

### Note also that a11 was changed after deepcopy was done

#### b11 is SHALLOW COPY, whereas c11 is DEEP COPY

#### we have not fully untangled the two objects (a11 and c11)

### problem: finding the square root of x (x > 0, can be any number -- not just an integer)

#### want to be within epsilon of the real square root

---

#### BRUTE FORCE ALGORITHM

---

```python
x = float(input("enter a number: "))
num_guesses = 0

ans = 0
epsilon = 0.00001

while ans **2 <= x:
    
    ans += epsilon
#    print(ans**2)
    num_guesses += 1

print(ans, "is close to the square root of ", x )
print(num_guesses, "is the number of guesses")
```

# Algorithms: Bisection Search

### setup a window which contains the square root

```python
x = float(input("enter a number: "))

epsilon = 0.00001
num_guesses = 0

low = 0
high = x

ans = (high + low)/2


# stop at the length of the window as 2 epsilon

while high-low >= 2 * epsilon:
    
    print("low = ", low, "high = ", high)
    num_guesses += 1
    
    if ans**2 < x:
        low = ans
    else:
        high = ans
    ans = (high + low) / 2


print(ans, "is close to the square root of x")
print(num_guesses, "is the number of guesses")

```

# Algorithms: Heron's Method

find ans close to the square root of x

consider x / ans

ans * (x/ans) = x

--- square root will be between ans and x/ans

next guess = (x/ans + ans) / 2

#### Heron's method is a special case of general purpose Newton's Method

-- when applied to a function

let f(ans) = ans**2 -x

find ans such that f(ans) = 0

<font color = red>Newton's methon when applied to the square root is the Heron's Method

```python
x = float(input("Enter a number: "))

epsilon = 0.00001
num_guesses = 0
ans = 1 # this is the first guess

while abs(x/ans - ans) > epsilon:
    ans = (x/ans + ans)/2
    num_guesses += 1

print(ans, 'is close to the square root of ', x)
print(num_guesses, "is the number of guesses")
```

<font size = 4, color = red>Class solution -- using prime = True & FOR Loop

```python
x = int(input("Enter a number :"))

prime = True

for item in range(2,x):
    
    # check if item divides x
    
    if x % item == 0:
        # now I know x is not a prime
        prime = False
if prime == True:
    print(x, "is a prime")
else:
    print(x, 'is not a prime')
        
```

```python
x = 109879

prime = True

for item in range(2,x):
    
    # check if item divides x
    print(x, item, x%item)
    if x % item == 0:
        
        # now I know x is not a prime
        prime = False
        break
if prime == True:
    print(x, "is a prime")
else:
    print(x, 'is not a prime')
```

<font size = 4, color = red>print all prime numbers from 1-100 & write to list --> (Class Solution using prime = True)

```python
mylist = []
for num in range(2,100):
    prime = True
    for item in range(2,num):
        if num % item == 0:
            prime = False
            break
    if prime == True:
#        print(num, "is a prime")
        mylist.append(num)
    else:
        pass
print(mylist)
```

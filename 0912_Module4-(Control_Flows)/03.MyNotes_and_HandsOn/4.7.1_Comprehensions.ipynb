### adding IF clause

```python
# unique characters in a word (in alphabetical order)
```

```python
import string

word = input("enter a word: ")

letters = [a for a in string.ascii_lowercase]

print(letters)
```

```python
# note this printed all letters in the alphabet
```

```python
import string

word = input("enter a word: ")

letters = [a for a in string.ascii_lowercase if a in word]

print(letters)
```

```python
# these are already in order

# letters are also unique
```

<font size = 5, color = blue>DICTIONARY COMPREHENSION

```python
# let us get a frequency count dictionary
```

```python
text = "Here is some example text"

frequencies = {letter: text.count(letter) for letter in text}

print(frequencies)
```

```python
# just my doodle (removing count)
```

```python
text = "Here is some example text"

frequencies = {letter: 'a' for letter in text}

print(frequencies)
```

<font size = 6, color = green>ADVANCED COMPREHENSIONS

```python
# formatting zip codes
```

```python
zip_list = [21122, '34454', '45556-2232', 56666, '90881-2345', 9089, 90098, 206]
```

```python
clean_zips = [x if type(x) == int else int(x[0:5]) for x in zip_list]

print(clean_zips)
```

---

```python
# modify list to 

zip_list2 = [21122, '34454', '45556-0232', 56666, '9088-0001', 9089, 90098, 206]
```

```python
clean_zips2 = [x if type(x) == int else int(x[0:5]) for x in zip_list2]

print(clean_zips2)
```

### this is the final working solution (from class)

```python
zip_list3 = [21122, '34454', '45556-0232', 56666, '9088-0001', 9089, 90098, 206]
```

```python
clean_zip3 = [int(x) if type(x) == int and len(str(x))==5 else int(x.split("-")[0])\
              if type(x) == str and len(x.split('-')[0]) ==5 else None for x in zip_list3]
```

```python
clean_zip3
```

```python
clean_zip4 = [int(x) if type(x) == int and len(str(x))==5 else int(x.split("-")[0])\
              if type(x) == str and len(x.split('-')[0]) ==5 else " " for x in zip_list3]
```

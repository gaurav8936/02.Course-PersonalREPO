# Titles, Ticks and Text

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
W = pd.read_csv('wages.csv')
```

```python
W = W.set_index('year')
```

```python
# color map takes values form 0 to 1 and 
colormap = plt.cm.gnuplot
```

```python
colors = [colormap(i) for i in np.linspace(0, .9, W.shape[1])]
```

```python
colors
```

---

```python
# dropping all workers
```

```python
W2 = W.drop('all_workers', axis = 1)
```

```python
W2.head()
```

```python
W2.plot(color = colors)
```

```python
# it has now selected colors from the colors list comprehension
```

```python
labeldict = {'less_than_high_school': 'less than high school',
            'high_school': 'high school',
             'some_college': 'some college',
             'college_degree': 'college degree',
             'advanced_degree': 'advanced degree'
            }
linestyledict = {'less_than_high_school': (0, (1,1)),
                'high_school': (0, (2,1)),
                 'some_college': (0, (4,1)),
                 'college_degree': (0, (6,1)),
                 'advanced_degree': (0, (10,1))
                }
colordict = dict(zip(['less_than_high_school',
            'high_school',
            'some_college',
             'college_degree',
             'advanced_degree'] , colors))
```

## Procedural Style

```python
for var in W2.columns:
    plt.plot(W2[var], label = labeldict[var],
            color = colordict[var],
            linestyle = linestyledict[var])
plt.legend()
```

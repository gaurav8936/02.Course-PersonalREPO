## sharex & sharey

```python
# how to compare heart and liver disease relate to alcohol
```

```python
fig, ax = plt.subplots(2,1)
ax[0].scatter(Wine.alcohol, Wine.heart)
ax[1].scatter(Wine.alcohol, Wine.liver)
```

---

```python
# line up axes

fig, ax = plt.subplots(2,1, sharex = True)
ax[0].scatter(Wine.alcohol, Wine.heart)
ax[1].scatter(Wine.alcohol, Wine.liver)
```

```python
# add title

fig, ax = plt.subplots(2,1, sharex = True)
ax[0].scatter(Wine.alcohol, Wine.heart)
ax[1].scatter(Wine.alcohol, Wine.liver)
ax[0].set_title("Heard Disease")
ax[1].set_title("Liver Disease")
```

```python
# fix spacing

fig, ax = plt.subplots(2,1, sharex = True)

fig.subplots_adjust(hspace =0.4)

ax[0].scatter(Wine.alcohol, Wine.heart)
ax[1].scatter(Wine.alcohol, Wine.liver)
ax[0].set_title("Heart Disease")
ax[1].set_title("Liver Disease")
```

---

```python
# add trend line

fig, ax = plt.subplots(2,1, sharex = True)

fig.subplots_adjust(hspace =0.4)

ax[0].scatter(Wine.alcohol, Wine.heart)
ax[1].scatter(Wine.alcohol, Wine.liver)
ax[0].set_title("Heard Disease")
ax[1].set_title("Liver Disease")
```

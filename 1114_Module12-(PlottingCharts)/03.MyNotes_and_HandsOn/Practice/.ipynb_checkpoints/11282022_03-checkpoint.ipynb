```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
W = pd.read_csv('wages.csv')
```

```python
W.shape
```

```python
W.head(5)
```

```python
W = W.set_index('year')
```

```python
W.head(5)
```

```python
# color map takes values form 0 to 1 and 
colormap = plt.cm.gnuplot
```

```python
colormap = plt.cm.tab20
```

```python
colors = [colormap(i) for i in np.linspace(0, .9, W.shape[1])]
```

```python
W2 = W.drop('all_workers', axis = 1)
```

```python
W2.head()
```

```python
W2.plot(color = colors)
```

```python
labeldict = {'less_than_high_school': 'less than high school',
            'high_school': 'high school',
             'some_college': 'some college',
             'college_degree': 'college degree',
             'advanced_degree': 'advanced degree'
            }
linestyledict = {'less_than_high_school': (0, (1,1)),
                'high_school': (0, (2,1)),
                 'some_college': (0, (4,1)),
                 'college_degree': (0, (6,1)),
                 'advanced_degree': (0, (10,1))
                }
colordict = dict(zip(['less_than_high_school',
            'high_school',
            'some_college',
             'college_degree',
             'advanced_degree'] , colors))
```

```python
for var in W2.columns:
    plt.plot(W2[var], label = labeldict[var],
            color = colordict[var],
            linestyle = linestyledict[var])
plt.legend()
```

---

```python
fig,ax = plt.subplots(1,1)

for var in W2.columns:
    ax.plot(W2[var], label = labeldict[var],
    color = colordict[var],
    linestyle = linestyledict[var])
ax.legend()
```

```python
fig,ax = plt.subplots(1,1)

for var in W2.columns:
    ax.plot(W2[var], label = labeldict[var],
    color = colordict[var],
    linestyle = linestyledict[var])
ax.set_ylim(0, 1600)
ax.legend()
```

```python
fig,ax = plt.subplots(1,1)

for var in W2.columns:
    ax.plot(W2[var], label = labeldict[var],
    color = colordict[var],
    linestyle = linestyledict[var])
ax.set_ylim(0, 1600)
ax.set_xticklabels(range(2000, 2019,2))
ax.legend()
```

```python
# add label for y axis
import matplotlib.ticker as mtick

fig, ax = plt.subplots(1,1, figsize = (10,5))
# fig = plt.plot(figsize = (20,5))

for var in W2.columns:
    ax.plot(W2[var], label = labeldict[var],
            color = colordict[var],
            linestyle = linestyledict[var])
    
ax.set_ylim(0, 1600)
ax.legend(ncol = 2, labelspacing = 0.1, borderaxespad = 0,  \
          bbox_to_anchor = (0.995, 0.01), loc = 'lower right')
ax.set_xticklabels(range(2000, 2019, 2))
ax.set_ylabel('weekly wage')

ax.set_title('Growth of Wage by Education Level')
ax.text(2008, 1100, 'housing crisis')

ax.annotate('look here', xy = (2011, W2.advanced_degree[2011]), 
           xytext = (2011, W2.advanced_degree[2011] + 200),
            
            arrowprops = {'facecolor': 'black'}
            
           )

fmt =  '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
pass 
```

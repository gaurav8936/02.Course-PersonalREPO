```python
pi = 3.1415926

print("{:.2f}".format(pi, precision))
```

```python
pi = 3.1415926

precision = 6

print("{:.{}f}".format(pi, precision))
```

```python
# pad number iwth left zeros

num = 5

width = 7

print("{:0>{}d}".format(num, width))
```

```python
# comma separator

num = 100000000

print("{:,}".format(num))
```

```python
num = 100000000

print("{:.2e}".format(num))
```

## Convert Values to Different Bases


A surprising use, you can use the string format command to convert numbers to different bases. Use the letter in the formatter to indicate which number base: decimal, hex, octal, or binary.



```python
print("{0:d} - {0:x} - {0:o} - {0:b}".format(21))
```

## Use Format as a Function


```python
email_f = "Your email address was {email}".format

print(email_f(email = "227monterey@gmail.com"))
```

---

Using format as a function can be used to adjust formating by user preference.



```python
num_format = "{:,}".format

print(num_format(1000000))
```

---

## Internationalization


To use locale specific formatting for numbers, you need to first set the locale, and then use the formating code n instead of d. For example, using commas or periods to separate thousands in numbers based on the users locale.



```python
import locale
locale.setlocale(locale.LC_ALL, '')

print("{:n}".format(1000000))
```

## Table Formatting Data


Use the width and the left and right justification to align your data into a nice table format. Heres an example to show how to format:



```python
starters = [
    [ 'Andre Iguodala', 4, 3, 7 ],
    [ 'Klay Thompson', 5, 0, 21 ],
    [ 'Stephen Curry', 5, 8, 36 ],
    [ 'Draymon Green', 9, 4, 11 ],
    [ 'Andrew Bogut', 3, 0, 2 ],
]

row = "| {player:<16s} | {reb:2d} | {ast:2d} | {pts:2d} |".format

for p in starters:
    print(row(player=p[0], reb=p[1], ast=p[2], pts=p[3]))
```

# F-Strings

```python
book = "Lord of the Rings"
author = "JRR Tolkein"

print(f"the {book} was written by {author}")
```

```python
# using the other format

print("the {book} was written by {author}".format(book = "Lord of the Rings", \
                                                  author = "JRR Tolkein"))
```

---

### 1. Do math with f-strings:



```python
print(f"do math: 3 * 6 = {3 * 6}")
```

### 2. Call functions with f-strings;



```python
verb = "runs"

print(f"The girl {verb.upper()} quickly")
```

*** You can use f-strings using the three different type of quotation marks in Python, single, double, or triple quotes. The following will all output the same:***

```python
name = "Fred"
print( f'{name}' )
print( f"{name}" )
print( f"""{name}""" )
```

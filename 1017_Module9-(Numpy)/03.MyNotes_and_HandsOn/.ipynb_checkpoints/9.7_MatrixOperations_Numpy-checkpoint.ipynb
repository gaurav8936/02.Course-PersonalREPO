# Matrix Operations in Numpy

```python
import numpy as np
```

```python
a= np.array([
    [1,2,3],
    [4,5,6]
    ])

```

```python
b = np.array([
    [1,1],
    [0,1],
    [2,2]
])
```

```python
a.shape, b.shape
```

```python
#because it is 3 X 3 for hte inner -- it works
```

```python
# old way to multiply
```

```python
a.dot(b)
```

```python
b.dot(a)
```

---

```python
a @ b
```

```python
b @ a
```

# Vectors

```python
# there is more than one way to store a vector in numpy
```

```python
v = np.array([1,2])
```

```python
v
```

```python
# this is a 1-dimensional array
```

```python
# is this a row or column vector
```

```python
# this is neither
```

```python
v.shape
```

```python
# this tells us only that this is a 1-dimensional array
```

---

```python
# to crate a row vector
```

```python
v_row = v.reshape(1,2)
```

```python
v_row
```

```python
# this is now a row vector
```

```python
v_row.shape
```

```python
# 1 row, 2 columns
```

---

```python
# pyt in a -1 insread of 2 --- that is the same effect 
# works for more generic scenarios (where you may not know 2)
```

```python
v_row = v.reshape(1,-1)
```

```python
v_row.shape
```

---

```python
v_col = v.reshape(-1, 1)
```

```python
v_col.shape
```

```python
v_col
```

```python
# this is a column vector
```

```python
# this is a standard way ot putting weights in a linear model
```

---

```python
b.shape, v_col.shape
```

```python
# we can now multiple these two
```

```python
b @ v_col
```

```python
# however the row version will not work
```

```python
b @ v_row
```

```python
b.shape, v_row.shape
```

---

```python
# remember the original vector v was just one-dimensional (neither row or column)
```

```python

b @ v
```

```python
b.shape, v.shape
```

```python
# so, in this case, python add 1 to the v.shape in order to allow the multiplication
```

---

<font size = 5, color = red> you should try to turn your one-dimensional vectors into column vectors and row vectors. And that's the way that you will very clearly know exactly what multiplication is happening.

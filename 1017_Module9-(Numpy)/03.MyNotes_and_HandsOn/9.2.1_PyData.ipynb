### Numpy supports a functional type of programming

    You'll probably find yourself programming in a functional style as you start using NumPy. 
    
    So for example, you'll start off with a matrix, an array of data, and then you pass that array through a sequence of functions. 
    
    And then that generates an output array. 
    
    And what's really important is that there is no state along the way. 
    
    So every time-- this sequence is completely reproducible. Every time you pass in the same array of data through these functions, the same exact thing will happen.
    
    And there's no side effects anywhere else in the program. 
    
    And so writing in this way can actually help make your code more transparent, can help communicate to the reader what it is that your code is trying to do.
